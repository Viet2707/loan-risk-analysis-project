```python
## 11. Feature grouping for analysis - Tạo nhóm dữ liệu mới
df["CreditScoreGroup"] = pd.cut(
    df["CreditScore"],
    bins=[300, 580, 670, 740, 800],
    labels=["Poor", "Fair", "Good", "Very Good"]
)
```

```python
df["IncomeGroup"] = pd.cut(
    df["AnnualIncome"],
    bins=[0, 30000, 60000, 100000, 500000],
    labels=["Low", "Medium", "High", "Very High"]
)
```

```python
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[18, 25, 35, 45, 55, 80],
    labels=["18-25", "26-35", "36-45", "46-55", "56-80"]
)
```
